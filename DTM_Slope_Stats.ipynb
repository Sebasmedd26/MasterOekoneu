```python
#import libraries

import rasterio as rs
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import math
from osgeo import gdal
from rasterio.plot import show
import os

print('*********** Libraries were imported successfuly **********')
print('working directory: '+ str(os.getcwd()))
```

```python
##Change directory to file folder
##os.chdir('D:\TEMP\20200324_Sentinel2A')

## open image to classify

gdal.UseExceptions()
img = rs.open ('DTM_Berlin_Okoneu_1M.tif')

print('**************** Image imported successfuly **************')

## Print image data
print('**********************************************************')
print('*********************** Image data ***********************')
print('Number of bands: ' + str(img.count))
print('Coordinate Reference System: ' + str(img.crs))
print('Image width:`' + str(img.width))
print('Image height:`' + str(img.height))
print('Number of Pixels:`' + str(int(img.height)*int(img.width)))
print('**********************************************************')
show(img)

## get bands
Band1 = img.read(1)

## stats
from scipy import stats
#stats.describe (Index_Values) #activate just if needed
```

```python
from rasterio.plot import show_hist
fig, axhist = plt.subplots(1, 1)
show_hist(img, bins=200, histtype='stepfilled',
          lw=0.0, stacked=False, alpha=0.6, ax=axhist, label="pixels")
axhist.set_xlabel("Datum (m)")
axhist.set_ylabel('Frequency')
axhist.set_title('Elevation Histogram')
plt.axvline(np.nanquantile(Band1, .25), 0, 7000000, label='Q1', ls="--")
plt.axvline(np.nanquantile(Band1, .75), 0, 7000000, label='Q3', ls="--")
```

```python
print("mean: " +str(np.nanmean(Band1)))
print("min: " +str(np.nanmin(Band1)))
print("max: " +str(np.nanmax(Band1)))
print("median: " +str(np.nanmedian(Band1)))
print("std: " +str(np.nanstd(Band1)))
print("CV: " + str("{0:.4f}".format(round((np.nanstd(Band1)/np.nanmean(Band1))*100,4))+" %"))
print("Q2 quantile of arr : ", np.nanquantile(Band1, .50)) 
print("Q1 quantile of arr : ", np.nanquantile(Band1, .25)) 
print("Q3 quantile of arr : ", np.nanquantile(Band1, .75)) 
print("100th quantile of arr : ", np.nanquantile(Band1, .1)) 
print("Q95 : ", np.nanquantile(Band1, .94)) 
```
